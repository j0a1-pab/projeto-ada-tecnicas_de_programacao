## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
faltantes = inv_df_by_id.isnull().sum()
print("~~~ Dados faltantes por coluna ~~~\n\n ---> absoluto:\n")
display(faltantes)
porcentagem = faltantes * 100 / len(inv_df_by_id)
print("\n\n ---> porcentagem:\
         \n (excluídas as colunas completas)\n")
for col, pct in porcentagem.items():
  if pct != 0: print(f"{col:<10} {pct:>10.2f}%")
```

```python
# Para tratar os dados faltantes, temos de pensar nas análises que faremos.
# Para os ativos, como se trata de valores investidos, faz sentido substituir por 0;
# afinal, podemos considerar que não ter dinheiro investido em dado ativo equivale
# a ter um investimento de 0 dinheiros. Assim a maioria das operações que fazem sentido
# para estes dados podem ser realizadas sem problemas (somas por investidor e por
# ativo, por exemplo, para comparar volumes de investimento; podemos ter problemas
# se quisermos utilizar médias ou identificar mínimos, mas nesses casos basta aplicar
# máscaras para filtrar e excluir os 0).

inv_df_by_id[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']] =\
 inv_df_by_id[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].fillna(0)

 # Obs.: para o método sum() NaN equivale a 0, por padrão nas versões mais recentes do numpy.
 # Já para somas diretas entre colunas (com o operador +) ou médias (mean()), o resultado
 # será NaN sempre que algum dos valores for NaN. Por isso é conveniente substituir por 0
 # (ou outro valor, como veremos abaixo). Além disso, a maior parte dos modelos de ML (da
 # biblioteca scikit-learn, por exemplo) não sabe lidar com NaN.
```

```python
# Já para as notas (NPS), trata-se de uma métrica cujo valor deve ser mantido, para
# cada investidorx, pois é característica de sua relação com a empresa (a agência
# de investimentos, no caso) -- e da evolução desta relação, entre uma avaliação
# e outra. Neste caso, substituir pelo valor 0 afetaria as medidas de tendência 
# central, tanto individuais quanto por período de avaliação. Até porque não está,
# a princípio, excluída a possibilidade de umx cliente de fato avaliar com nota 0...

# Poderíamos pensar em duplicar a nota existente, nos casos em que x cliente/investidorx
# tenha feito apenas uma avaliação. Isso, porém, não resolveria o problema, pois,
# como podemos observar pela saída abaixo, há várixs clientes (165) que não fizeram
# nenhuma das duas avaliações.

inv_df_by_id[(inv_df_by_id['Nota1'].isna()) & (inv_df_by_id['Nota2'].isna())][['Nota1', 'Nota2']]
```

```python
# Escolhemos, então, substituir os dados faltantes nas colunas Nota1 e Nota2 por -1
# (pois as notas são via de regra valores não negativos). Assim, ao fazer nossas
# análises, bastará tomar o cuidado de filtrar esses valores negativos.

inv_df_by_id[['Nota1', 'Nota2']] =\
 inv_df_by_id[['Nota1', 'Nota2']].fillna(-1)
```

```python
# agora não temos mais valores nulos no DF:
display(inv_df_by_id.isnull().sum().any())
```

```python
# vizualizando as notas faltantes:
inv_df_by_id[(inv_df_by_id['Nota1'] == -1) | (inv_df_by_id['Nota2'] == -1)][['Nota1', 'Nota2']]
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
df = inv_df_by_id.copy()
df.loc[:, 'Total'] = df['Ativo01'] + df['Ativo02']\
 + df['Ativo03'] + df['Ativo04'] + df['Ativo05'] + df['Ativo06'] + df['Ativo07']

# Reordenando as colunas, para que o Total apareça logo depois dos ativos:
df.insert(9, 'Total', df.pop('Total'))

df['Total'].head(3)
```

```python
# Para a média das notas vou definir uma função, para tomar a média apenas quando houver duas notas
# (caso haja uma com valor -1, a média será igual à outra nota; caso ambas sejam -1, será -1)

def media(x, y):
  if (x != -1) and (y != -1):
    return (x + y) / 2
  elif x == -1:
    return y
  else:
    return x

df['Media'] = df.apply(lambda x: media(x['Nota1'], x['Nota2']), axis=1)
df.insert(12, 'Media', df.pop('Media'))

df['Media'].head(3)
```

```python
# Dominio:

df['Dominio'] = df['Email'].apply(lambda x: x.split('@')[1])
df['Dominio'].head(3)
```

```python
# Renda:

df = df.rename(columns = {'Salario':'Renda'})
df['Renda'] = df['Renda'].apply(lambda x: float(x[1:]))
df['Renda'].head(3)
```

```python
# Mudar os dados da coluna "Sexo" para Masculino e Feminino (usando o map):

df['Sexo'] = df['Sexo'].map(lambda x: 'Feminino' if x == 'Fem' else 'Masculino')
df['Sexo'].head(3)
```

```python
# Transformar os dados da coluna Media em inteiro:
display(df['Media'].head(3))
df['Media'] = df['Media'].map(lambda x: int(round(x)))
df['Media'].head(3)
```

```python
# "NPS":

def nps(x):
  if x >= 80:
    return 'promotor'
  elif x >= 50:
    return 'neutro'
  else:
    return 'detrator'

df['NPS'] = df['Media'].apply(nps)
df['NPS'].head(3)
```
